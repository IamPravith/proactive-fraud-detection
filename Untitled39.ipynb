# FRAUD DETECTION MODEL

## 1. IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

## 2. LOAD DATA

```python
df = pd.read_csv( r"C:\Users\Pravith Kumar J\Downloads\Fraud.csv",low_memory=False)    
```

```python
print("Dataset Shape:", df.shape)
print("\nFraud Label Check:")
print(df['isFraud'].value_counts())
```

## 3. DATA CLEANING

## Identify numeric & categorical columns

```python
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns
```

## Handle missing values


```python
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

## Outlier treatment 

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower, upper)
```

## 4. MULTICOLLINEARITY CHECK 

```python
X_vif = df.select_dtypes(include=np.number).drop(
    columns=['isFraud', 'isFlaggedFraud'],
    errors='ignore'
)
X_vif = X_vif.loc[:, X_vif.nunique() > 1]

vif = pd.DataFrame()
vif["Feature"] = X_vif.columns
vif["VIF"] = [
    variance_inflation_factor(X_vif.values, i)
    for i in range(X_vif.shape[1])
]
```

```python
print("\nVIF TABLE\n", vif)
```

## 5. FEATURE PREPARATION

## Drop ID columns

```python
df = df.drop(columns=['nameOrig', 'nameDest'])
```

## Encode categorical column

```python
df = pd.get_dummies(df, columns=['type'], drop_first=True)
```

## Features only (unsupervised)

```python
X = df.drop(columns=['isFraud', 'isFlaggedFraud'], errors='ignore')
```

## 6. CALIBRATION VALIDATION SPLIT

```python
X_train, X_val = train_test_split(X, test_size=0.2, random_state=42)
```

## 7. FEATURE SCALING

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

## 8. FRAUD DETECTION MODEL- ISOLATION FOREST

```python
iso_model = IsolationForest(
    n_estimators=300,
    contamination=0.001,   # assumed fraud rate
    random_state=42,
    n_jobs=-1
)

iso_model.fit(X_train_scaled)
```

## 9. VALIDATION PERFORMANCE

```python
val_scores = iso_model.decision_function(X_val_scaled)
val_preds = iso_model.predict(X_val_scaled)
```

## Convert predictions: -1 → fraud, 1 → normal

```python
val_preds = pd.Series(val_preds).map({-1: 1, 1: 0})
```

```python
print("\nValidation Potential Fraud Count:")
print(val_preds.value_counts())
```

```python
print("\nValidation Anomaly Score Summary:")
print(pd.Series(val_scores).describe())
```

## 10. APPLYING MODEL TO FULL DATA

```python
X_full_scaled = scaler.transform(X)
df['anomaly_score'] = iso_model.decision_function(X_full_scaled)
df['potential_fraud'] = iso_model.predict(X_full_scaled)
df['potential_fraud'] = df['potential_fraud'].map({-1: 1, 1: 0})
```

## 11. BUSINESS METRICS

```python
total_txns = len(df)
flagged_txns = df['potential_fraud'].sum()

print("\nBUSINESS KPIs")
print("Total Transactions:", total_txns)
print("Flagged Suspicious Transactions:", flagged_txns)
print("Estimated Fraud Rate:", flagged_txns / total_txns)
```

## 12. VISUALIZATION

```python
plt.figure(figsize=(6,4))
sns.countplot(
    data=df,
    x='potential_fraud',
    hue='potential_fraud',
    palette={0: 'lightgreen', 1: 'red'},
    legend=False
)
plt.title("Normal vs Potential Fraud Transactions")
plt.xlabel("0 = Normal | 1 = Potential Fraud")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(
    val_scores,
    bins=50,
    color='red'
)
plt.title("Validation Anomaly Score Distribution")
plt.xlabel("Anomaly Score")
plt.show()
```
